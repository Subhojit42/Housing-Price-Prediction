### RESCALING THE FEATURES
As you saw in the demonstration for simple linear regression, scaling doesn't impact your model. Here we can see that except
for area , all the columns have small integer values . so it is extremely important to rescale the variables so that they have
a comparable scale. if we don't  have comparable scales, then some of the other coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coeffients. this might become very annoying at the time of model evaluation. so it is advised to use standardisation or normalization so that the units of the coefficients obtained are all on the same scale. as you know there are two common ways of rescaling.
    1. Min-Max scaling (normalization)
    2. Standardization (mean=0, sigma=1)
This time , we will use Min-Max scaling.

```python
# x
# normalization= (x-xmin)/(xmax-xmin)
# standarization= (x-mu)/sigma
```

```python
# instantiate the object
scaler = MinMaxScaler()

# create the list of numeric variable
num_vars= ['area' , 'bedrooms', 'bathrooms', 'stories', 
          'parking' , 'price']

# fit on data
    # fit(): learns xmin, xmax
    # transform(): x-xmin/xmax-xmin
    #fit_transform()
    
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

##  CHECKING VIF
    - variation inflation factor or vif gives us a basic quantitative idea about how much the feature variables are correlated with each other . it is extremely important parameter to our linear model. the formula for calculating VIF is:
    
    VIF = 1/(1-r^2)

```python
# create a dataframe that will contain the names of all the feature variables and their respect.

vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values , i)  for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF" , ascending= False)
vif
```

we could have: 
- high p-value , high VIF: remove them
- high-low:
    - high p, low VIF: remove these first
    - low p, high VIF: remove these after the first ones above
- low p, low VIF: keep them

```python
X = X_train.drop('semi-furnished' , axis=1)
```

```python
# bulid another model with all the variables after removing semi furnished
X_train_sm_new = sm.add_constant(X)

# create the first model
lr = sm.OLS(y_train, X_train_sm_new)

# fit
lr_model = lr.fit()

# parameters
lr_model.params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values , i)  for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF" , ascending= False)
vif
```

```python
X_1 = X.drop('bedrooms', axis=1)
```

```python
# bulid another model with all the variables after removing semi furnished
X_train_sm_new_1 = sm.add_constant(X_1)

# create the first model
lr = sm.OLS(y_train, X_train_sm_new_1)

# fit
lr_model = lr.fit()

# parameters
lr_model.params
lr_model.summary()
```

```python
vif = pd.DataFrame()
vif['Features'] = X_1.columns
vif['VIF'] = [variance_inflation_factor(X_1.values , i)  for i in range(X_1.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF" , ascending= False)
vif
```

## step 4: Predictions and evaluations on the test set
